```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dtaset
file_path = "Heart_Disease_Prediction.csv"
heart_df = pd.read_csv(file_path)

# Check dimensions
print("Step 1: Check Dimensions")
print(f"Dimensions of the dataset: {heart_df.shape}")
print("\n" + "="*50 + "\n")

# Summary statistics
print("Step 2: Summary Statistics")
print(heart_df.describe())
print("\n" + "="*50 + "\n")

# Data types
print("Step 3: Data Types")
print(heart_df.dtypes)
print("\n" + "="*50 + "\n")

# Display the columns in the DataFrame
print("Step 4: Original Columns")
print(heart_df.columns)
print("\n" + "="*50 + "\n")

# Strip leading/trailing spaces from column names
heart_df.columns = heart_df.columns.str.strip()

# Check column names again
print("Step 5: Stripped Columns")
print(heart_df.columns)
print("\n" + "="*50 + "\n")

# Rename the 'Heart Disease' column to 'target'
heart_df.rename(columns={'Heart Disease': 'target'}, inplace=True)

# Verify if 'target' column exists
print("Step 6: Verify 'target' Column")
if 'target' in heart_df.columns:
    print("Column 'target' exists.")
else:
    print("Column 'target' does NOT exist.")
print("\n" + "="*50 + "\n")
plt.show()
```

```python
#BAR PLOT
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x='target', data=heart_df)
plt.title('Distribution of Heart Disease')
plt.show()
```

```python
#HISTOGRAMS
heart_df.hist(figsize=(20, 10))
plt.suptitle('Histograms of Heart Disease Features')
plt.show()
```

```python
# Display the columns in the DataFrame
print("Columns in DataFrame:")
print(heart_df.columns)
```

```python

# Bar plot for 'Chest pain type'
plt.figure(figsize=(10, 6))
heart_df['Chest pain type'].value_counts().plot(kind='bar', color='darkgreen')
plt.title('Counts of Each Chest Pain Type')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot for 'Age' vs 'Cholesterol'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Cholesterol', y='Age', data=heart_df, color='green')
plt.title('Age vs Cholesterol')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.show()
```

```python
# Bar plot - Top categories in 'Chest pain type'
plt.figure(figsize=(10, 6))
top_chest_pain_types = heart_df['Chest pain type'].value_counts().head(10)
top_chest_pain_types.plot(kind='barh', color='darkblue')
plt.title('Top Categories in Chest Pain Type')
plt.xlabel('Number of Patients')
plt.ylabel('Chest Pain Type')
plt.tight_layout()
plt.show()
```

```python
# Histogram - Distribution of Cholesterol levels
plt.figure(figsize=(7, 6))
sns.histplot(heart_df['Cholesterol'], bins=40, kde=True, color='red')
plt.title('Distribution of Cholesterol Levels')
plt.xlabel('Cholesterol')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Box plot for Age by Chest Pain Type
plt.figure(figsize=(10, 6))
sns.boxplot(x='Chest pain type', y='Age', data=heart_df, palette='pastel')
plt.title('Age Distribution by Chest Pain Type')
plt.xlabel('Chest Pain Type')
plt.ylabel('Age')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
